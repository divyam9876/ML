```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
wine=pd.read_csv("mlfolder/CSV_Files/wine.csv")
```

```python
wine.head()
```

```python
acidity=wine.iloc[:,0]
density=wine.iloc[:,4]
```

```python
fig, ax=plt.subplots()
fig.suptitle("fixed_acidity versus density")
ax.set_xlabel("fixed_acidity")
ax.set_ylabel("density")
ax.scatter(acidity,density,color='blue')
plt.show()
```

```python
x1=np.linspace(0,2*np.pi,100)
y1=np.sin(x1)
y2=np.cos(x1)
```

```python
fig, ax= plt.subplots()
fig.suptitle("Sine and Cosine")
ax.plot(x1,y1,color='red',label='sinx')
ax.plot(x1,y2,color='blue',label='cosx')
ax.legend(loc='upper center')
plt.show()
fig.set_figheight(8)
fig.set_figwidth(8)
```

```python
fig, ax=plt.subplots(3,2)
fig.suptitle("Some Trig Functios",fontweight='bold')
ax[0][0].set_title("tangent")
ax[0][1].set_title('cotangent')
ax[1][0].set_title('sin')
ax[1][1].set_title('cosec')
ax[2][0].set_title('cos')
ax[2][1].set_title('sec')
fig.set_figheight(8)
fig.set_figwidth(10)
plt.tight_layout(pad=3)
plt.show()
```
