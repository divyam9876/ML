```python
wineData=pd.read_csv("wine_small.csv")
```

```python
wineData
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = wineData.iloc[:,:-1]
y=wineData.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
Xtrain,Xtest,yTrain,yTest = train_test_split(X,y,random_state=42,test_size=0.25)
```

```python
yTest
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(Xtrain,yTrain)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
prediction=reg.predict(Xtest)
```

```python
newData=pd.DataFrame([[.58,11]])
```

```python
reg.predict(newData)
```

```python
reg.score(Xtest,yTest)
```

```python
reg.score(Xtrain,yTrain)
```
