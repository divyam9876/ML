### (c) Shuffle the data (use random seed 12)

```python
Data=sonarData.sample(frac=1, random_state=12).reset_index(drop=True)
```

### (d)Split the data into two sets, one called train_data consisting of 80% of all data, and the remaining in the other set called test_data. Use subsetting to do this.

```python
length_of_data= len(Data)
```

```python
train_size = int(length_of_data*.8)
```

```python
train_size
```

```python
test_size = 208-train_size
test_size
```

#### We need to split the last column, column 61, of the trainData to get x_train and y_train.

#### After doing this, x_train should have a size of 166x60 and y_train should be of size 166x1.

#### Similarly, we need to split the last column, column 61, of the testData to get x_test and y_test.

#### After doing this, x_test should have a size of 42x60 and y_train should be of size 42x1


```python
trainData =  Data.iloc[:train_size,:]
testData = Data.iloc[train_size:,:]
```

```python
## Separate out the last column of training data
x_train = trainData.iloc[:,:60]
y_train = trainData.iloc[:,60]

## Separate out the last column of test data
x_test = testData.iloc[:,:60]
y_test = testData.iloc[:,60]
```
