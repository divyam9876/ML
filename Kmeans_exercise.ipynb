```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("mlfolder/CSV_Files/scatterdata.csv")
```

```python
data.head(3)
```

```python
data2=data.iloc[:,:2]
```

```python
#initial centroids
start=np.array([[100,100],[400,200]])
```

```python
cluster_centers=[start]
for i in range(6):
    km=KMeans(n_clusters=2,max_iter=i+1,init=start,random_state=42)
    km.fit(data2)
    cluster_centers.append(km.cluster_centers_)
```

```python
cluster_centers
```

```python
a=[]
b=[]
for i in cluster_centers:
    a.append(i[[0][0]])
for i in cluster_centers:
    b.append(i[[1][0]])
```

```python
a
```

```python
b
```

```python
arr_width=2
fig,ax=plt.subplots(1,1,figsize=(10,8))
ax.scatter(data2.iloc[:,0],data2.iloc[:,1],c=km.labels_,cmap='winter')
for i in range(5):
    ax.arrow(a[i][0],a[i][1], a[i+1][0]-a[i][0], a[i+1][1]-a[i][1],head_width =5 * arr_width, 
          head_length = 6* arr_width, color='b', lw=1, length_includes_head=True, zorder=1)
    ax.arrow(b[i][0],b[i][1], b[i+1][0]-b[i][0], b[i+1][1]-b[i][1],head_width = 5 * arr_width, 
          head_length = 5 * arr_width, color='r', lw=1, length_includes_head=True, zorder=1)
```
