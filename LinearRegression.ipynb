```python
data2=pd.read_csv("mlfolder/CSV_Files/regression_example_data.csv")
```

```python
data2.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data2.iloc[:,0]
y=data2.iloc[:,1]
```

```python
X=X.to_frame()
y=y.to_frame()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=142,shuffle=True)
```

```python
model2=LinearRegression()
```

```python
model2.fit(X_train,y_train)
```

```python
a=model2.intercept_
b=model2.coef_
a=a.item()
b=b.item()
print('a:',a)
print('b:',b)
```

```python
prediction=model2.predict(X_test)
```

```python
prediction=pd.DataFrame(prediction)
```

```python
fig,ax=plt.subplots()
ax.set_xlabel('X')
ax.set_ylabel('y')
ax.scatter(X,y,color='blue',label='data')
ax.plot(X,a+b*X,color='red',label='RegressionLine')
ax.legend()
```

```python
fig,ax=plt.subplots()
ax.set_xlabel('X')
ax.set_ylabel('y')
ax.scatter(X_train,y_train,color='green', label='training data')
ax.scatter(X_test,prediction,color='Maroon', label='predicted test data')
ax.scatter(X_test,y_test,color='blue',label='test data')
ax.plot(X_train,a+b*X_train,color='Red', label='Training data Regression Line')
# ax.plot(X_test,a+b*X_test,color='brown',label='Test data Regression Line')
ax.legend()
fig.set_figheight(5)
fig.set_figwidth(8)
plt.show()
```

```python
model2.score(X_test,prediction)
```
