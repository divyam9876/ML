```python
#gradient descent
'''
x0 ← some initial guess of x
M ← max iterations
epsilon ← tolerance, a very small positive value
alpha ← step size, or learning rate
i ← int =1
f(x) ← given function
f’(x) ← first derivative of f(x) with respect to x, df/dx
while (i < M):
xi+1 ← xi – alpha * f’(xi) (update xi by going down the slope)
if abs(xi+1 – xi) < epsilon: (if x is not changing, stop)
break
i++
return xi (this is where we found that f(x) achieves its minimum.
An alternative stopping criterion for the iteration will be when the function f(xi) doesn’t change too
much:
If abs(f(xi+1) – f(xi)) < epsilon, break.
'''
```

```python
def costFunction(t0, t1, x, y):
    return 1/(2*len(x))* np.sum( np.power((t0+t1*x-y), 2))
```

```python
def gradientDescent(t0 , t1 , x, y, alpha, maxIter=10000, tolerance=.00001):
    '''
    Computes the regression coefficients using gradient descent for a single indepenedent variable.
    t0 and t1 are initial values of regression coefficicents, x, y are input data, alpha is learning 
    rate, maxIter is maximum iterations done, tolerance is  error tolerance (we stop iteration if the 
    successive values of t0 and t1 differ by less than tolerance)
    '''
    ## initialize a (maxIter, 2) shaped 2D array to hold theta values (not an efficient thing to do)
    theta=np.zeros((maxIter,2))
    ## initialize a (maxIter,) shaped 1D array to hold J values (not an efficient thing to do). J represents the error function. 
    ## J is what we are trying to minimize
    J=np.zeros((maxIter,))
    m=len(x)
    for i in range(maxIter):
        ''' 
        Compute new values of t0 and t1, say t0_new and t1_new, and store them in the theta array
        Remember you are doing the i-th iteration here so the values produced should go to the i-th rows of  
        theta and J arrays.
        '''
        t0New=t0-alpha/m*np.sum(t0+t1*x-y)  # new t0  after going down the slope      
        t1New=t1-alpha/m*np.sum((t0+t1*x-y)*x)   # new t1  after going down the slope
        theta[i,0]=t0New # save updated t0
        theta[i,1]=t1New # save updated t1
        t0Er=np.abs(t0-t0New) # compute how much t0 changed in this iteration
        t1Er=np.abs(t1-t1New) # compute how much t1 changed in this iteration 
        t0=t0New ## update t0
        t1=t1New ## update t1
        ## Compute cost function J, store its value in the appropriate location in the  J array  
        J[i]=costFunction(t0, t1, x, y) ## calculate ans save cost J
        # check for convergence - quit if errors do not change to within tolerance
        if (t0Er<tolerance) and (t1Er<tolerance):    
            print("Iteration converged at i=", i)  # print out when we stopped iteration
            break
    return theta, J
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Step 1: Generate random 2D data points
rng = np.random.default_rng(seed=42)
X = np.dot(rng.random((2, 2)), rng.standard_normal((2, 200))).T  # Random 2D data

# Step 2: Apply PCA to the data
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 3: Plot the original data and principal components
plt.figure(figsize=(8, 6))

# Original data
plt.scatter(X[:, 0], X[:, 1], alpha=0.6, label="Original Data", c='gray')

# Principal components
origin = np.mean(X, axis=0)  # The center of the data
for i, vector in enumerate(pca.components_):
    plt.arrow(origin[0], origin[1],
              vector[0] * 2 * np.sqrt(pca.explained_variance_[i]),
              vector[1] * 2 * np.sqrt(pca.explained_variance_[i]),
              color=f'C{i}', width=0.02, label=f'PC{i+1}')

# Labels and title
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("PCA Visualization")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid(True)
plt.show()

# Step 4: Explained variance
print("Explained Variance Ratios:", pca.explained_variance_ratio_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import matplotlib.animation as animation

# Step 1: Generate random 2D data points
rng = np.random.default_rng(seed=42)
X = np.dot(rng.random((2, 2)), rng.standard_normal((2, 200))).T  # Random 2D data

# Step 2: Apply PCA to the data
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 3: Setup for dynamic visualization
fig, ax = plt.subplots(figsize=(8, 6))

# Original data points
ax.scatter(X[:, 0], X[:, 1], alpha=0.6, label="Original Data", c='blue')

# Labels and title
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_title("Rotating Principal Components")
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
ax.axvline(0, color='black', linewidth=0.5, linestyle='--')
ax.legend()

# Calculate the origin (mean of data)
origin = np.mean(X, axis=0)

# Initialize arrows for PC1 and PC2
pc1_arrow = ax.annotate('', xy=(0, 0), xytext=origin, arrowprops=dict(color='C0', width=1))
pc2_arrow = ax.annotate('', xy=(0, 0), xytext=origin, arrowprops=dict(color='C1', width=1))

# Function to update the plot for animation
def update(frame):
    angle = np.radians(frame)
    rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)],
                                [np.sin(angle),  np.cos(angle)]])
    rotated_components = np.dot(rotation_matrix, pca.components_)

    # Update PC1
    pc1_arrow.set_position((origin[0] + rotated_components[0, 0] * 2 * np.sqrt(pca.explained_variance_[0]),
                            origin[1] + rotated_components[0, 1] * 2 * np.sqrt(pca.explained_variance_[0])))
    # Update PC2
    pc2_arrow.set_position((origin[0] + rotated_components[1, 0] * 2 * np.sqrt(pca.explained_variance_[1]),
                            origin[1] + rotated_components[1, 1] * 2 * np.sqrt(pca.explained_variance_[1])))

# Animate principal components
ani = animation.FuncAnimation(fig, update, frames=np.arange(0, 360, 2), interval=50)

plt.show()

# Step 4: Explained variance
print("Explained Variance Ratios:", pca.explained_variance_ratio_)
```

**Outliers Cleaning**

```python
def outlierCleaner(y_predicted, x, y):
    """
        Delete 10% of points that have the largest
        residual errors (difference between the prediction
        and the actual y).
        Return a list of tuples named cleaned_data where 
        each tuple is of the form (x, y, error).
    """
    
    cleaned_data = []
    import numpy as np
    error = np.abs(y-y_predicted)
    n=int(len(error)*0.1)
    ind=np.argpartition(error.ravel(),-n)[-n:]
    x_clean=np.delete(x,ind)
    y_clean=np.delete(y,ind)
    err=np.delete(error,ind)
    cleaned_data=zip(x_clean,y_clean,err)
    return cleaned_data
```

**KMeans**

```python
import random
import math

def euclidean_distance(x1, x2):
    return math.sqrt(sum((x1[i] - x2[i]) ** 2 for i in range(len(x1))))

def k_means(X, K, max_iterations=100, tolerance=1e-4):
    centroids = random.sample(X, K)
    old_centroids = [[0] * len(centroids[0])] * len(centroids)

    for _ in range(max_iterations):
        closest_centroids = [0] * len(X)
        
        for i in range(len(X)):
            distances = [euclidean_distance(X[i], centroid) for centroid in centroids]
            closest_centroids[i] = distances.index(min(distances))

        new_centroids = []
        for k in range(K):
            cluster_points = [X[i] for i in range(len(X)) if closest_centroids[i] == k]
            if cluster_points:
                new_centroid = [sum(point[i] for point in cluster_points) / len(cluster_points) for i in range(len(X[0]))]
                new_centroids.append(new_centroid)
            else:
                new_centroids.append(centroids[k])
        
        diff = sum(math.sqrt(sum((new_centroids[i][j] - old_centroids[i][j]) ** 2 for j in range(len(new_centroids[i])))) for i in range(K))
        if diff < tolerance:
            break
        
        old_centroids = new_centroids
    
    return new_centroids, closest_centroids

X = [[1, 2], [1, 4], [1, 0], [10, 2], [10, 4]]
K = 2
centroids, cluster_assignments = k_means(X, K)

print("Centroids:", centroids)
print("Cluster assignments:", cluster_assignments)
```
