```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
train=np.genfromtxt('mlfolder/CSV_Files/data_for_SVM_exercises/trainData.csv',delimiter=',',skip_header=True)
```

```python
test=np.genfromtxt('mlfolder/CSV_Files/data_for_SVM_exercises/testData.csv',delimiter=',',skip_header=True)
```

```python
X_train=train[:,np.arange(2)]
```

```python
y_train=train[:,2]
```

```python
X_test=test[:,np.arange(2)]
y_test=test[:,2]
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_train,cmap='winter')
```

```python
from sklearn.svm import SVC
```

```python
model1=SVC(kernel='rbf',C=10000,gamma='scale')
model1.fit(X_train,y_train)
```

```python
model1.predict(X_test)
```

```python
np.sum(model1.predict(X_test)==y_test)/len(y_test)
```

**Plotting Decision Boundary**

```python
h=0.01
xmin=0
ymin=0
xmax=1
ymax=1
xx,yy=np.meshgrid(np.arange(xmin,xmax,h),np.arange(ymin,ymax,h))
Z=model1.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
plt.contourf(xx,yy,Z,cmap='winter')
plt.scatter(X_test[:,0],X_test[:,1],c=y_test,s=50,cmap='autumn')
plt.text(.7,.8,'yellow and red dots \n are test data')
plt.show()
```
