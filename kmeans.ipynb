```python
d=np.genfromtxt('mlfolder/CSV_Files/biCluster.csv',delimiter=',',skip_header=1)
```

```python
df=pd.DataFrame(d)
```

```python
sns.scatterplot(data=df,x=df.iloc[:,0],y=df.iloc[:,1],legend=False)
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
X=df.iloc[:,np.arange(2)]
```

```python
X.shape
```

```python
kmeans=KMeans(n_clusters=2,random_state=0).fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
plt.scatter(X.iloc[:,0],X.iloc[:,1],c=kmeans.labels_,s=50, cmap='viridis')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1], c='red', s=200, alpha=0.7)
```

```python
kmeans = KMeans(n_clusters=2, init=np.array([[ 3.37548103,  3.14213765],[ 2.20353872,  6.37112557]]), random_state=0, max_iter=1).fit(X)
centers = kmeans.cluster_centers_
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.7)
```

```python
kmeans = KMeans(n_clusters=2, init=np.array([[ 3.37548103,  3.14213765],[ 2.20353872,  6.37112557]]), random_state=0, max_iter=6).fit(X)
centers = kmeans.cluster_centers_
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.7)
```
