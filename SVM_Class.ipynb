```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
data=np.genfromtxt('mlfolder/CSV_Files/linearlyseparable.csv')
```

```python
data.shape
```

```python
X=data[:,:2] #X=data[:,range(2)]
```

```python
y=data[:,2]
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
```

```python
from sklearn.svm import SVC
```

```python
model1=SVC(kernel='linear',C=1E10)
```

```python
model1.fit(X,y)
```

```python
model1.coef_
```

```python
model1.intercept_
```

```python
np.sum(model1.predict(X)!=y)
```

**Accuracy is 100%**

**Plotting Decision Boundary**

```python
h=0.01
xmin=-1
xmax=4
ymin=-1
ymax=6
xx,yy=np.meshgrid(np.arange(xmin,xmax,h),np.arange(ymin,ymax,h))
Z=model1.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
plt.scatter(X[:,0],X[:,1],c=y,s=70,cmap='winter')#s=70 size of points
plt.contour(xx,yy,Z)
plt.show()
```

**plt.contour(xx,yy,Z)** not filled areas
**plt.contourdf(xx,yy,Z)** filled areas with colors

# NON LINEAR SVM

**kernel='polynomial','radial'**

```python
data2=np.genfromtxt('mlfolder/CSV_Files/non-linearly-separable.csv')
```

```python
data.shape
```

```python
X2=data2[:,range(2)]
y2=data2[:,2]
```

```python
plt.scatter(X2[:,0],X2[:,1],c=y2,cmap='winter')
```

```python
model2=SVC(kernel='rbf',C=1)
```

```python
model2.fit(X2,y2)
```

```python
np.sum(y2!=model2.predict(X2))
```

```python
#plotting Decision Boundary
h=0.01
xmin=-1.5
xmax=1.5
ymin=-1.5
ymax=1.5
xx,yy=np.meshgrid(np.arange(xmin,xmax,h),np.arange(ymin,ymax,h))
Z=model2.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
plt.contour(xx,yy,Z)
plt.scatter(X2[:,0],X2[:,1],c=y2,cmap='winter')
plt.show()
```
