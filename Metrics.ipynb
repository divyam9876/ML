```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,1,200)
df=pd.DataFrame({'Precision=1':2*x*1/(x+1),
                'Precision=0.75':2*x*0.75/(x+0.75),
                'Precision=0.5':2*x*0.5/(x+0.5),
                'Precision=0.25':2*x*0.25/(x+0.25),
                'Precision=0.1':2*x*0.1/(x+0.1)})
plt.plot(x,'Precision=1',data=df,color='blue',linewidth=3)
plt.plot(x,'Precision=0.75',data=df,color='purple',linewidth=3)
plt.plot(x,'Precision=0.5',data=df,color='violet',linewidth=3)
plt.plot(x,'Precision=0.25',data=df,color='orange',linewidth=3)
plt.plot(x,'Precision=0.1',data=df,color='red',linewidth=3)
plt.xlabel('Recall')
plt.ylabel('F1')
plt.legend()
```
